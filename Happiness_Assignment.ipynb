```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data= pd.read_csv('/content/combined (World Happiness Data)_combined (1).csv', sep='\t')
```

```python
data.head()
```

```python
data.info()
```

```python
data['Year'] = pd.to_datetime(data['Year'].astype(str), format='%Y')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.drop('reset filters', axis=1, inplace=True)
data.head()
```

```python
# Set up the overall style for plots
sns.set(style="whitegrid")

# Plot: Happiness Trends Over Time by Region (with no shaded region)
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Happiness Score', hue='Region', marker='o', errorbar=None)

# Customize the plot
plt.title('Happiness Score Trends Over Time by Region', fontsize=16, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
#Interactive plot
fig = px.line(data, x="Year", y="Happiness Score", color='Region', title='Happiness Score Trends Over Time by Region')
fig.show()
```

From the graph, we observe that the Australia and New Zealand region consistently maintains a high happiness score, averaging around 7.3 throughout the years. This stability over the years suggests that the region’s happiness could be supported by certain factors like strong economic performance, accessible healthcare, and high levels of personal freedom and social support.

In contrast, the Sub-Saharan African region shows a significantly lower average happiness score of 4.2 in 2015. Unlike Australia and New Zealand, Sub-Saharan Africa experiences noticeable fluctuations in happiness scores over time. The region saw a slight dip in happiness between 2016 and 2017, possibly reflecting things like economic hardships, political instability, or social issues impacting happiness. However, after 2017, the happiness score in this region experienced a gradual increase, reaching its highest point of 4.4 in 2019. This upward trend could indicate improvements in things like economic conditions or healthcare systems across parts of the region.



```python
top_10_happiest= data.sort_values(by=[ 'Happiness Score'], ascending=[False])
fig = px.bar(top_10_happiest.head(45), x="Region", y="Happiness Score", color="Country", title='Top 10 Happiest Countries by Region')
fig.show()
```

```python
# Setting the figure size and theme
plt.figure(figsize=(10, 6))

# Customizing the box plot
sns.boxplot(x='Region', y='Happiness Score', data=data, palette="viridis", linewidth=1.2, hue='Region', legend=False)

# Adding titles and labels with larger fonts
plt.title('Distribution of Happiness Score by Region', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

# Rotating x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Displaying the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick labels
plt.show()
```

The box plot shows the distribution of happiness scores according to each region. We can see that in Western Europe, the happiness score ranges between 4.9 to 7.8 and that of Australia and New Zealand ranges between 6.9 and 7.3. Even though Western Europe has several high-ranking countries, such as Finland and Denmark, other Western European countries score lower than these top countries, which impacts the regional average and makes it appear lower than Australia and New Zealand in the regional ranking of happiness score.

1.**Finland**-Finland's happiness score gradually increases between 2015 and 2019. The rise in Finland’s happiness score during this period reflects the impact of its well-rounded social welfare system, high levels of trust and cohesion, and emphasis on personal well-being and environmental health. Finland’s commitment to work-life balance, quality education, mental health, and environmental sustainability created a stable and supportive environment that allowed its citizens to thrive, leading to consistently high levels of happiness and life satisfaction.

2.**Denmark**-Denmark's happiness score was stable between 2015 and 2017 but after 2017 it gradually rose up to 2019. Denmark’s happiness score remained stable between 2015 and 2017 due to a solid foundation of economic stability, social trust, and work-life balance. From 2017 to 2019, the gradual rise in happiness was driven by a combination of economic growth, mental health initiatives, enhanced community engagement, and environmental and cultural policies that promoted both personal and community well-being. These factors reinforced the country’s quality of life and supported a further increase in happiness levels among Danish citizens.

3.**Switzerland**-Switzerland has been experiencing a gradual decrease in its happiness score over time. Switzerland’s gradual decrease in happiness scores from 2015 to 2019 reflects the influence of high living costs, financial stress, mental health challenges, climate concerns, and high social expectations. While Switzerland’s social systems and infrastructure remain strong, these subtle but significant pressures have impacted people’s day-to-day experiences, reducing overall happiness.

4.**Iceland**-Iceland also experienced a decrease in its happiness score over time. Iceland’s gradual decrease in happiness scores between 2015 and 2019 can be attributed to the pressures of rising living costs, housing affordability challenges, economic dependence on tourism, and increasing mental health issues. Additionally, the strain on the environment due to tourism, social isolation, and political uncertainty contributed to a decline in overall happiness. Despite the country's high standard of living, these underlying issues led to stress and dissatisfaction, which lowered happiness levels over time.

```python
least_happy_countries = ['Central African Republic', 'Togo', 'South Sudan', 'Burundi', 'Syria']
filtered_data = data[data['Country'].isin(least_happy_countries)]
fig = px.line(filtered_data, x="Year", y="Happiness Score", color='Country')
fig.show()
```

1.**C.A.R** - C.A.R's happiness score fell from 2015 and reached an all time low in 2017. It had been having civil wars from 2013 causing political instability in the area. Following years of severe conflict, the international community increased its humanitarian aid to C.A.R. in an effort to stabilize the region. This led to an increase in happiness score after 2017. Furthermore, After 2017, efforts to establish peace agreements gained traction, with several rounds of peace talks involving the government and armed groups. In early 2019, the African Union and United Nations brokered a peace agreement in Khartoum, signed by the government and 14 armed groups.

2.**Togo** - Even though Togo ranks low in happiness score, its happiness has been gradually increasing over the years. This might be attributed to Togo's consistent economic growth during this period,improved access to education and healthcare which led to Infant mortality drop and more children completing school compared to other years.

3.**South Sudan**- South sudan's happiness score has been gradually decreasing since 2016 to 2019. This drop can be attributed to political instability. The country has been facing ongoing civil conflict since its independence in 2011, with a severe escalation between 2013 and 2018. The prolonged conflict led to one of the world’s largest displacement crises. Furthermore, the country experienced severe food insecurity, with several regions on the brink of famine during this period.

4.**Burundi**-Burundi’s happiness score remained stagnant from 2015 to 2018 due to ongoing political, social, and economic challenges that hindered improvement in quality of life. In 2015, Burundi experienced a severe political crisis when then-president Pierre Nkurunziza announced he would seek a third term, sparking protests and an attempted coup. This period saw significant political repression, violence, and social instability. Starting 2018, some small but important shifts allowed for a gradual improvement in the happiness score. By 2018, political tensions began to ease as the government worked to stabilize the country . Moreover, the East African Community (EAC) and other regional bodies began to play a more active role in supporting peace and stability in Burundi, facilitating diplomatic efforts to ease political tensions.

5.**Syria**-Syria's happiness score increases gradually from 2015 up to 2017 then plateaus up to 2019. This is explained by the intensity of the conflict in Syria decreasing slightly during this period. In addition to that, Syria received international humanitarian aid and reconstruction efforts in some regions which caused the increased happiness score. Furthermore, the people of Syria had adapted psychologically to challenging conditions, leading to gradual increases in reported well-being. Between 2017 and 2019 it’s possible that improvements reached a ceiling. Any gains made through reduced violence or aid might not have extended further, leading to a stabilization in happiness levels in this period.

##INTERSECTION EXPLANATION
Burundi's happiness score has remained largely stagnant due to ongoing political instability and economic challenges that stifle significant progress in well-being. On the other hand, Togo has seen a steady improvement, driven by economic reforms and relative political stability, allowing it to move up from one of the lowest positions since 2015. This growth led to an intersection point with Burundi's score as Togo caught up through consistent social and governance improvements.







```python
least_happy_countries = ['Central African Republic', 'Togo', 'South Sudan', 'Burundi', 'Syria']
filtered_data = data[data['Country'].isin(least_happy_countries)]
columns_of_interest = ['Happiness Score', 'Perception of Corruption', 'Economy (GDP per Capita)',
                       'Health (Life Expectancy)', 'Freedom to make life choices', 'Generosity',
                       'Social support']
filtered_data = filtered_data[columns_of_interest]
correlation_matrix = filtered_data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Factors Affecting Happiness in the Least Happy Countries')
plt.show()
```

Countries with high GDP have higher social support and high haappiness scores as well.
